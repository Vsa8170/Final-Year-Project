#importing libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#importing the dataset


```python
dataset = pd.read_csv('Training.csv')
df_train = pd.read_csv('Training.csv', delimiter=',')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
k=y
diseases=y.reshape(len(y),1)
symptoms=np.array(df_train.columns)
print(x.shape)
```

#Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
# diseases = np.array(ct.fit_transform(diseases))
```

#Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 0)
```

```python
print(y_train)
```

```python
print(y_test)
```

#Training the K-NN model on the Training set


```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

#Predicting a new result

```python
description = np.array(input("Describe your symptoms: ").split())
```

```python
s=np.zeros(132)
i=0
for x in symptoms:
   if x in  description:
      s[i]=1
   i+=1
```

```python
B = s.reshape(1,132)
print(B)
```

```python
print(classifier.predict(B))
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred)*100)
```
